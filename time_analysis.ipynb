# Análises do tempo de processamento

## Importa as bibliotecas necessárias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Plota o gráfico

```python
# Read data
df = pd.read_csv('etl/log.txt', header=None, sep=' ')

# Plot lines
sns.lineplot(data=df[0], label='Time')

# X label
plt.xlabel('Leitura')
# Y label
plt.ylabel('Tempo (ms)')
# Title
plt.title('Tempo de leitura do arquivo de log')
```
